```python
import numpy as np
import cv2
import os


def loadimages(path = "."):
    return [f for f in sorted(os.listdir(path)) if f.endswith('.png')]

index = 0
width = 224
height = 224
dim = (width, height)

path1= path + 'image_segmentation/AugFr/'
path2= path + 'image_segmentation/AugMasks/'

filenames=loadimages(path1)
filemasks=loadimages(path2)

for file in filenames:
    
    image = cv2.imread(path1 + file )
    mask = cv2.imread(path2 + 'Mask'+  file)
    resizedFr = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
    resizedM = cv2.resize(mask, dim, interpolation = cv2.INTER_AREA)
  
    cv2.imwrite(path + 'image_segmentation/ResizedFr/'+ file, resizedFr)
    cv2.imwrite(path + 'image_segmentation/ResizedMasks/'+ 'Mask' + file, resizedM)
```
