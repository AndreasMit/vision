```python
import numpy as np
import cv2
import os


def loadimages(path = "."):
    return [f for f in sorted(os.listdir(path)) if f.endswith('.png')]


path1 = path + 'image_segmentation/ResizedMasks/'
filenames=loadimages(path1)

for file in filenames:
    mask = cv2.imread(path1 + file )
    CMask = (mask/255).astype('uint8')
    cv2.imwrite(path + 'image_segmentation/CMasks/'+ file, CMask)
```

```python
print("After saving:")  
print(os.listdir( path + 'image_segmentation/CMasks'))
```

```python
path2 = path + 'image_segmentation/ResizedFr/'
path3 = path + 'image_segmentation/CMasks/'

frames = loadimages(path2)
masks = loadimages(path3)
```

```python
#random choose 10% of them and move them to testFr and testM
import random
n = int(0.1 * np.shape(frames)[0])
print(n)#number of test frames
test = random.sample(frames, n)
```

```python
import shutil
for t in test:
    src = path + 'image_segmentation/ResizedFr/'+ t
    dst = path + 'image_segmentation/TestFr/'+ t
    shutil.move(src, dst)
```

```python
for t in test:
    src = path + 'image_segmentation/CMasks/Mask'+ t
    dst = path + 'image_segmentation/TestM/'+ t
    shutil.move(src, dst)
```
